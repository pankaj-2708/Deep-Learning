```python
# removing aposhtrophes

contractions = [
    "i'm",
    "you're",
    "he's",
    "she's",
    "it's",
    "we're",
    "they're",
    "i've",
    "you've",
    "we've",
    "they've",
    "i'd",
    "you'd",
    "he'd",
    "she'd",
    "we'd",
    "they'd",
    "i'll",
    "you'll",
    "he'll",
    "she'll",
    "we'll",
    "they'll",
    "isn't",
    "aren't",
    "wasn't",
    "weren't",
    "haven't",
    "hasn't",
    "hadn't",
    "don't",
    "doesn't",
    "didn't",
    "can't",
    "couldn't",
    "won't",
    "wouldn't",
    "shouldn't",
    "mustn't",
    "what's",
    "where's",
    "when's",
    "who's",
    "how's",
    "that's",
    "there's",
    "y'all"
]

explanations=['i am',
 'you are',
 'he is ',
 'she is ',
 'it is ',
 'we are',
 'they are',
 'i have',
 'you have',
 'we have',
 'they have',
 'i would ',
 'you would ',
 'he would ',
 'she would ',
 'we would ',
 'they would ',
 'i will',
 'you will',
 'he will',
 'she will',
 'we will',
 'they will',
 'is not',
 'are not',
 'was not',
 'were not',
 'have not',
 'has not',
 'had not',
 'do not',
 'does not',
 'did not',
 'cannot',
 'could not',
 'will not',
 'would not',
 'should not',
 'must not',
 'what is ',
 'where is ',
 'when is ',
 'who is ',
 'how is ',
 'that is',
 'there is ',
 'you all']

units_dict = {
    # Base units
    "m": "metre",
    "kg": "kilogram",
    "s": "second",
    "A": "ampere",
    "K": "kelvin",
    "mol": "mole",
    "cd": "candela",

    # Derived units
    "Hz": "hertz",
    "N": "newton",
    "Pa": "pascal",
    "J": "joule",
    "W": "watt",
    "C": "coulomb",
    "V": "volt",
    "Ω": "ohm",
    "S": "siemens",
    "F": "farad",
    "T": "tesla",
    "H": "henry",
    "lm": "lumen",
    "lx": "lux",

    # Everyday usage
    "km": "kilometre",
    "cm": "centimetre",
    "mm": "millimetre",
    "µm": "micrometre",
    "nm": "nanometre",
    "km/h": "kilometre per hour",
    "m/s": "metre per second",
    "L": "litre",
    "ml": "millilitre",
    "g": "gram",
    "mg": "milligram",
    "°C": "degree Celsius",
    "°F": "degree Fahrenheit",
    "%": "percent",

    # Non-SI but common
    "min": "minute",
    "h": "hour",
    "d": "day",
    "°": "degree",
    "rad": "radian",
    "sr": "steradian",
    "bar": "bar",
    "atm": "atmosphere",
    "cal": "calorie",
    "eV": "electronvolt"
}
```

```python
# lowercase
# stripping
# removing links 
# removing email
# there is latex written in abstract for now i am removing it 
# symbol replacement with letter like % , e
# removing #,>,",',(,)
# expanding units like km/h 

# stopword removal
# stemming
import string
import regex as re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
import nltk

nltk.download("stopwords")
stpWrd = stopwords.words("english")

punc=string.punctuation
def preprocess_text(txt):
    txt.strip().lower().replace("%","percent")
    # e expansion 
    pat=" e "
    re.sub(pat," exponential ",txt)
    # for email removal
    pat="[\w+._%+-]+@[\w+._%+-]+\.[\w+]{2,}"
    txt=re.sub(pat,"",txt)
    # for links
    pat="https?://\S+"
    txt=re.sub(pat,"",txt)
    # for latex
    pat="\$\$.*?\$\$"
    re.sub(pat,"",txt)
    # removing punc
    for i in punc:
        txt.replace(i,"")
    # apostrophes expansion
    for i in range(len(contractions)):
        txt.replace(contractions[i],explanations[i])
    # units expansion
    for i in range(len(units_dict)):
        txt.replace(list(units_dict.keys())[i],list(units_dict.values())[i])
    # stemming and stopword removal
    ps = PorterStemmer()
    new_txt = []
    for word in txt.split():
        if word not in stpWrd and word.isalnum():
            new_txt.append(ps.stem(word))

    return " ".join(new_txt)
```

```python
from tqdm import tqdm
tqdm.pandas()
```

```python
df['abstract']=df['abstract'].progress_apply(preprocess_text)
df['title']=df['title'].progress_apply(preprocess_text)
df['venue']=df['venue'].progress_apply(preprocess_text)
```
