```python
# converting an rgb image to grayscale image

img = cv2.imread('./training_set/training_set/cats/cat.1.jpg', 1)
print("rgb image shape",img.shape)

img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print("grayscale image shape",img_gray.shape)
img_gray
```

```python
# reading diff channels of rgb image
img = cv2.imread('./training_set/training_set/cats/cat.1.jpg', 1)
img_blue=img[:,:,0]
img_green=img[:,:,1]
img_red=img[:,:,2]
np.hstack((img_blue,img_green,img_red))
```

```python
# altering channels
img1 = cv2.imread('./training_set/training_set/cats/cat.1.jpg', 1)
img2 = cv2.imread('./training_set/training_set/cats/cat.1.jpg', 1)
img3 = cv2.imread('./training_set/training_set/cats/cat.1.jpg', 1)

# removing blue color
img1[:,:,0]=0

# removing green color
img2[:,:,1]=0

# removing red color
img3[:,:,2]=0

np.hstack((img1,img2,img3))
```

```python
# resizing image
img=cv2.imread('./training_set/training_set/cats/cat.1.jpg', 1)
print(img.shape)
img_resize=cv2.resize(img, (200,200))
print(img_resize.shape)
img_resize
```

```python
# flipping image
img=cv2.imread('./training_set/training_set/cats/cat.1.jpg', 1)

img_flip_ver=cv2.flip(img, 0)

img_flip_hor=cv2.flip(img, 1)

img_flip_both=cv2.flip(img, -1)

np.hstack((img_flip_ver,img_flip_hor,img_flip_both))
```

```python
# drawing rectangle on image
img=cv2.imread('./training_set/training_set/cats/cat.1.jpg', 1)
# to draw rectangle we have to pass location of diagonally opp pt of rectangle
# if thickness=-1 then rectangle will be filled
cv2.rectangle(img, (100,100), (200,200), color=(0,0,255), thichness=2)
img
```

```python
# drawing circle on image
img=cv2.imread('./training_set/training_set/cats/cat.1.jpg', 1)
# for drawing circle we have to pass location of centre and its radius
# if thickness=-1 then circle will be filled
cv2.circle(img, (100,100), 50, color=(0,0,255), thichness=2)
img
```

```python
# drawing line on image
img=cv2.imread('./training_set/training_set/cats/cat.1.jpg', 1)
# for drawing line we have to pass start and end point of line
cv2.line(img, (100,100), (200,200), color=(0,0,255), thichness=2)
img
```

```python
# writting text on image
img=cv2.imread('./training_set/training_set/cats/cat.1.jpg', 1)
# for text we have to pass start location
cv2.putText(img, text='cat', (100,100), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=2, color=(0,0,255), thickness=2)
img
```
